```python
############################################################
# Solution.py
# Author: Jagadeesh Vasudevamurthy
# Copyright: Jagadeesh Vasudevamurthy 2025
###########################################################

############################################################
# All imports
###########################################################
#from Util import *
import heapq

############################################################
# NOTHING CAN BE CHANGED in Solution
###########################################################
class Solution:
    def __init__(self,a:'list of runtime', num_races: 'list of size 1', r:'list of 3 best students', show:'bool'):
        self.N = 25 # Number of students
        self.S = 3 # First 3 winners
        self.T = 5 # Number of tracks

        self._a = a # List of run time (Given)
        self._num_races = num_races ; # MUST fill number of races conducted
        self._r = r #Find first 3 students (Must fill)
        self._show = show # Must print how you conducted races is _show is True
        ## All action happens in class Alg
        a = Alg(self,show)

    def fill_rank(self, student_number:'int', rank:'int'):
        assert(student_number >= 0 and student_number <= self.N)
        assert(rank >= 0 and rank <= self.S-1)
        self._r[rank] = student_number

    def update_num_races(self, n:'int'):
        self._num_races[0] = self._num_races[0] + n

    def get_student_runtime(self,student_number:'int')->'int':
        assert(student_number >= 0 and student_number <= self.N)
        return self._a[student_number]

############################################################
# You will write code in class Alg
# YOU CAN HAVE ANY NUMBER OF CLASSES 
###########################################################
class Alg:
    def __init__(self,s:'Solution',show:'bool'):
        self._s = s
        self._show = show
        ## YOU CAN HAVE ANY NUMBER OF PRIVATE VARIABLES
        
        ## All happens in _race
        self._race()
        

    ############################################################
    # WRITE YOUR CODE BELOW
    # YOU CAN HAVE ANY NUMBER OF CLASSES AND PRIVATE FUNCTION
    # if self._show = True MUST SHOW ALL RACES
    ###########################################################
    def _race(self):
        # Divide the students into 5 groups
        # Conduct a race for each group
        if (self._show): print("------------------------------------------------------------Conducting Race 1: ")
        race1 = self._conduct_race([0, 1, 2, 3, 4])
        if (self._show): print("------------------------------------------------------------Conducting Race 2: ")
        race2 = self._conduct_race([5, 6, 7, 8, 9])
        if (self._show): print("------------------------------------------------------------Conducting Race 3: ")
        race3 = self._conduct_race([10, 11, 12, 13, 14])
        if (self._show): print("------------------------------------------------------------Conducting Race 4: ")
        race4 = self._conduct_race([15, 16, 17, 18, 19])
        if (self._show): print("------------------------------------------------------------Conducting Race 5: ")
        race5 = self._conduct_race([20, 21, 22, 23, 24])
        
        # Conduct a race amongst the 1st place winners of each group
        if (self._show): print("------------------------------------------------------------Conducting Race 6: ")
        race6 = self._conduct_race6(race1, race2, race3, race4, race5)

        # Consider runtime of winner of the students in raceA > raceB > raceC
        raceA = heapq.heappop(race6).get_element()
        raceB = heapq.heappop(race6).get_element()
        raceC = heapq.heappop(race6).get_element()

        # Ulitamate first place winner is the winner of raceA (race6)
        first_winner = heapq.heappop(raceA).get_element()
        if (self._show): print("------------------------------------------------------------First Winner: Student ID:", first_winner)
        self._s.fill_rank(first_winner, 0)

        # Now the only thing left is to conduct a race to determine the 2nd and 3rd place winners
        # The only possibilities are : 
        # C1 may be third 
        # B1 may be second or third 
        # B2 may be third 
        # A2 may be second or third 
        # A3 may be third 
        A2_student_number = heapq.heappop(raceA).get_element()
        A3_student_number = heapq.heappop(raceA).get_element()
        B1_student_number = heapq.heappop(raceB).get_element()
        B2_student_number = heapq.heappop(raceB).get_element()
        C1_student_number = heapq.heappop(raceC).get_element()

        if (self._show): print("------------------------------------------------------------Conducting Race 7: ")
        race7 = self._conduct_race([A2_student_number, A3_student_number, B1_student_number, B2_student_number, C1_student_number])

        seccond_winner = heapq.heappop(race7).get_element()
        if (self._show): print("------------------------------------------------------------Seccond Winner: Student ID:", seccond_winner)
        self._s.fill_rank(seccond_winner, 1)

        third_winner = heapq.heappop(race7).get_element()
        if (self._show): print("------------------------------------------------------------Third Winner: Student ID:", third_winner)
        self._s.fill_rank(third_winner, 2)   


        if (self._show): 
            print("Top 3 for tournamanet: ")
            print("I: ", first_winner)
            print("II: ", seccond_winner)
            print("III: ", third_winner)
            print("Total number of races: ", self._s._num_races) 

    def _conduct_race(self, student_list):
        race = []
        race_copy = []
        for i in range(len(student_list)):
            student_number = student_list[i]
            student_runtime = self._s.get_student_runtime(student_number)
            if (self._show): print("Track Number: ", i,  " | Student number: ", student_number, " | Runtime: ", student_runtime)
            heapq.heappush(race, PQ_Element(student_runtime, student_number))
            heapq.heappush(race_copy, PQ_Element(student_runtime, student_number))
        self._s.update_num_races(1)
        if (self._show): 
            print("I : ", heapq.heappop(race_copy).get_element())
            print("II : ", heapq.heappop(race_copy).get_element())
            print("III: ", heapq.heappop(race_copy).get_element())
        return race               
        
    def _conduct_race6(self, race1, race2, race3, race4, race5):
        self._conduct_race([race1[0].get_element(), race2[0].get_element(), race3[0].get_element(), race4[0].get_element(), race5[0].get_element()])
        race6 = []
        heapq.heappush(race6, self._create_race6_ref(race1))
        heapq.heappush(race6, self._create_race6_ref(race2))
        heapq.heappush(race6, self._create_race6_ref(race3))
        heapq.heappush(race6, self._create_race6_ref(race4))
        heapq.heappush(race6, self._create_race6_ref(race5))
        return race6

    def _create_race6_ref(self, race):
        race_pq_winner = race[0]
        student_number = race_pq_winner.get_element()
        return PQ_Element(self._s.get_student_runtime(student_number), race)
        

        
        
class PQ_Element:
    def __init__(self, run_time: 'int', element):
        self._run_time = run_time
        self._element = element

    def __lt__(self, rhs: 'PQ_Element') -> 'bool':
        if (self._run_time < rhs._run_time):
            return True
        return False
    
    def __str__(self):
        return "(" + str(self._run_time) + " " + str(self._element) + ")"
    
    def __repr__(self):
        return self.__str__()
    
    def get_element(self):
        return self._element
```
